```python
#ignore Warnings
import warnings
warnings.filterwarnings("ignore")

# keras imports
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.optimizers import SGD
from keras.constraints import maxnorm
from keras.utils import np_utils

#import dataset
from keras.datasets import cifar10

#labels for images
labels = ['Airplane', 'Automobile', 'Bird',
         'Cat', 'Deer', 'Dog', 'Frog', 'Horse',
         'Skip', 'Truck']


#train/test
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

new_X_train = X_train.astype('float32')
new_X_test = X_test.astype('float32')
new_X_train /= 255
new_X_test /= 255
new_y_train = np_utils.to_categorical(y_train)
new_y_test = np_utils.to_categorical(y_test)

model = Sequential()
model.add(Conv2D(32,                     #insure orientation wont effect
                  (3, 3),                # model.
                  input_shape=(32, 32, 3),
                  activation='relu',
                  padding='same',
                  kernel_constraint=maxnorm(3))) 

#dont ignore any part of the image
model.add(MaxPooling2D(pool_size=(2,2))) 

model.add(Flatten()) #set up for Dense & Dropout

#which characteristics make the picture
model.add(Dense(512,
                  activation='relu',
                  kernel_constraint=maxnorm(3)))

#ensure that we are not going into too much detail and model
#doesnt get too lazy
model.add(Dropout(0.5))

#extra degree of accuracy
model.add(Dense(10, activation='softmax'))

#what kind of loss function
model.compile(loss='categorical_crossentropy',
              optimizer=SGD(lr=0.01),
              metrics=['accuracy'])

#fit model
model.fit(new_X_train, new_y_train, epochs=100, batch_size=32)

#save model
import h5py
model.save('Trained_model.h5')
```
